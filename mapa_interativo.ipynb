```python
import folium
import requests
from bs4 import BeautifulSoup
from geopy.geocoders import Nominatim
from unidecode import unidecode


geolocator = Nominatim(user_agent='pertin-alpha')


class Lugar():
    #um objeto pra estruturar os dados retornados do geopy/Nominatim
    def __init__(self, local):
        lugarPesquisado = geolocator.geocode(local, addressdetails=True)
        if lugarPesquisado == None:
            self.naoExiste = 1
        else: 
            self.naoExiste = 0
            self.lat = lugarPesquisado.raw['lat']
            self.lon = lugarPesquisado.raw['lon']
            self.addr = lugarPesquisado.raw['address']
            self.estado = limpador(lugarPesquisado.raw['address']['state'])
            if self.estado == 'rio-de-janeiro':
                self.abrev = 'rj'
            if self.estado == 'sao-paulo':
                self.abrev = 'sp'
            try:
                self.cidade = limpador(lugarPesquisado.raw['address']['town'])
            except:
                self.cidade = limpador(lugarPesquisado.raw['address']['city'])
            try:
                self.bairro = limpador(lugarPesquisado.raw['address']['city_district'])
            except:
                try:
                    self.bairro = limpador(lugarPesquisado.raw['address']['suburb'])
                except:
                    self.bairro = None

def limpador(palavra):
    #autoexplicativo
    palavra = unidecode(palavra.lower())
    palavra = palavra.replace(' ', '-')
    return palavra


def webScrapper(localidade):
    #aqui ele vai pegar o local que voce quer morar perto e pesquisar os ap's no bairro dele
    localidade = Lugar(localidade)
    if localidade.naoExiste == 1:
        return None
    pagina = requests.get(f'https://www.vivareal.com.br/aluguel/{localidade.estado}/{localidade.cidade}/bairros/{localidade.bairro}/apartamento_residencial/')
    sopa = BeautifulSoup(pagina.text, 'html.parser')
    if sopa.find(class_ = 'error__title').text != None or sopa.find(class_ = 'results-summary__count js-total-records').text == 0:
        pagina = requests.get(f'https://www.vivareal.com.br/aluguel/{localidade.estado}/{localidade.cidade}/apartamento_residencial/')
        sopa = BeautifulSoup(pagina.text, 'html.parser')
        if sopa.find(class_ = 'error__title').text != None or sopa.find(class_ = 'results-summary__count js-total-records').text == 0:
            pagina = requests.get(f'https://www.vivareal.com.br/aluguel/{localidade.abrev}/{localidade.cidade}/apartamento_residencial/')
            sopa = BeautifulSoup(pagina.text, 'html.parser')
            if sopa.find(class_ = 'error__title').text != None or sopa.find(class_ = 'results-summary__count js-total-records').text == 0:
                pagina = requests.get(f'https://www.vivareal.com.br/aluguel/{localidade.abrev}/{localidade.cidade}/apartamento_residencial/')
                sopa = BeautifulSoup(pagina.text, 'html.parser')
  
    endereco = sopa.find_all(class_='property-card__address')
    for end in endereco:
        geo = Lugar(end)
        if geo != None:
            end = geo

    #print(endereco)
    '''else:
        pagina = requests.get(f'https://www.vivareal.com.br/aluguel/{dados_local[0]}/{dados_local[1]}/bairros/{dados_local[2]}/apartamento_residencial/?pagina={indice}')
    sopa = BeautifulSoup(pagina.text, 'html.parser')
    endereco = sopa.find_all(class_='property-card__address')
    sopinha = sopa.find_all(class_ = 'js-change-page')
    paginas = []
    for legumes in sopinha:
        try:
            paginas.append(int(legumes.text))
        except:
            pass
    try:
        indice = max(paginas)+1
    except:
        pass
    for i in range(2,indice+1):
        try:    
            if dados_local[2] == 'vazio':
                pagina = requests.get(f'https://www.vivareal.com.br/aluguel/{dados_local[0]}/{dados_local[1]}/apartamento_residencial/?pagina={i}')
            else:
                pagina = requests.get(f'https://www.vivareal.com.br/aluguel/{dados_local[0]}/{dados_local[1]}/bairros/{dados_local[2]}/apartamento_residencial/?pagina={i}')
            sopa = BeautifulSoup(pagina.text, 'html.parser')
            for elementos in sopa.find_all(class_='property-card__address'):
                endereco.append(elementos)
        except:
            pass
    ender = []
    for ennndd in enumerate(endereco):
        endereco[ennndd[0]] = endereco[ennndd[0]].text
    print(endereco)
    for j in endereco:
        try:
            for k in range(len(endereco)):
                if j == endereco[k]:
                    endereco.pop(k)
        except:
            pass
    print(endereco)
    print(len(endereco))
    for enderecos in endereco:
        asdf = geolocator.geocode(enderecos, addressdetails=True)
        try: 
            ender.append([asdf.raw['lat'], asdf.raw['lon']])
        except:
            pass

    endereco = list(dict.fromkeys(endereco))
'''
objetivo = 'ufjf'
webScrapper(objetivo)
```
